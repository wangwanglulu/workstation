## Linear Equation Systems
$Ax =b$ where A is a m × n matrix, b is a m × 1 matrix (or m-vector), and
x is the unknown n × 1 solution matrix (or n-vector).

```python
from scipy import linalg as la
from scipy import optimize
import sympy
sympy.init_printing()
import numpy as np
import matplotlib.pyplot as plt
```

A system for which $n < m$ is said to be underdetermined (欠定), because it has fewer
equations than unknown and therefore cannot completely determine a unique solution.
If, on the other hand, $m > n$, then the equations are said to be overdetermined（超定）. This will
in general lead to conflicting constraints, resulting in that a solution does not exist.

$m = n$ is an important special case (unique solution). the
matrix $A$ must be nonsingular（非奇异）, in which case the inverse of $A$ exists, and the solution can
be written as $x = A^{−1}b$. If the matrix $A$ is singular, that is, the rank of the matrix （秩） is less than
$n$, $rank(A) < n$, or, equivalently, if its determinant（行列式） is zero, $detA = 0$, then the equation
$Ax = b$ can either have no solution or infinitely many solutions, depending on the right-hand
side vector $b$.

The advantage of using SymPy is of course that we may obtain exact results and
we can also include symbolic variables in the matrices.
$$\left(
\begin{matrix}
1 & \sqrt{p}\\
1 & 1/\sqrt{p}
\end{matrix}
\right)
\left(
\begin{matrix}
x_1\\
x_2
\end{matrix}
\right)=\left(
\begin{matrix}
1\\
2
\end{matrix}
\right)
$$

```python
p = sympy.symbols("p", positive=True)
A = sympy.Matrix([[1, sympy.sqrt(p)], 
                  [1, 1/sympy.sqrt(p)]])
b = sympy.Matrix([1, 2])
x = A.solve(b);x
```

Underdetermined systems have more variables than equations
$$\left(
\begin{matrix}
1 & 2 & 3\\
4 & 5 & 6
\end{matrix}
\right)
\left(
\begin{matrix}
x_1\\
x_2\\
x_3
\end{matrix}
\right)=\left(
\begin{matrix}
7\\
8
\end{matrix}
\right)
$$

```python
x_vars = sympy.symbols("x_1, x_2, x_3")
A = sympy.Matrix([[1, 2, 3], [4, 5, 6]])
x = sympy.Matrix(x_vars)
b = sympy.Matrix([7, 8])
sympy.solve(A*x - b, x_vars)
```

If the system is overdetermined and has more equations than
unknown variables, $m > n$, then we have more constraints than degrees of freedom, and in general there is no exact solution to such a system.

$$\left(
\begin{matrix}
1 & x_1 & x^2_1\\
\vdots & \vdots & \vdots \\
1 & x_m & x^2_m
\end{matrix}
\right)
\left(
\begin{matrix}
A\\
B\\
C
\end{matrix}
\right)=\left(
\begin{matrix}
y_1\\
y_2\\
y_3
\end{matrix}
\right)
$$

```python
# minimize the sum of square errors
# true model
x = np.linspace(-1, 1, 100)
a, b, c = 1, 2, 3
y_exact = a + b * x + c * x**2

#noisy data
m = 100
X = 1 - 2 * np.random.rand(m)
Y = a + b * X + c * X**2 + np.random.randn(m)
```

```python
A = np.vstack([X**0, X**1, X**2])
sol, r, rank, sv = la.lstsq(A.T, Y)
# r 残差项，rank 矩阵的秩，sv 奇异值
```

```python
y_fit = sol[0] + sol[1] * x + sol[2] * x**2
fig, ax = plt.subplots(figsize=(12, 3))
ax.plot(X, Y, 'go', alpha=0.5, label='Simulated data')
ax.plot(x, y_exact, 'k', lw=2, 
        label='True value $y = 1 + 2x +3x^2$')
ax.plot(x, y_fit, 'b', lw=2, label='Least square fit')
ax.set_xlabel(r"$x$", fontsize=18)
ax.set_ylabel(r"$y$", fontsize=18)
ax.legend(loc=2)        
```

linear model vs. higher-order polynomial model (overfit).
Using an appropriate model is an important and delicate aspect of
data fitting.

## Nonlinear Equations
A nonlinear equation can always be written on the form $f (x) = 0$, where $f (x)$ is a nonlinear function.

```python
### Univariate Equations
x, a, b, c = sympy.symbols("x, a, b, c")
sympy.solve(a + b*x + c*x**2, x)
```

However, in general nonlinear equations are typically not solvable analytically.

```python
x = np.linspace(-2, 2, 1000)
f1 = x**2 - x - 1
f2 = x**3 - 3 * np.sin(x)
f3 = np.exp(x) - 2
f4 = 1 - x**2 + np.sin(50 / (1 + x**2))
```

```python
fig, axes = plt.subplots(1, 4, figsize=(12, 2), 
                         sharey=True)
for n, f in enumerate([f1, f2, f3, f4]):
    axes[n].plot(x, f, lw=1.5)
    axes[n].axhline(0, ls=':', color='k')
    axes[n].set_ylim(-5, 5)
    axes[n].set_xticks([-2, -1, 0, 1, 2])
    axes[n].set_xlabel(r'$x$', fontsize=18)
    axes[0].set_ylabel(r'$f(x)$', fontsize=18)
```

### Bisection method

a starting interval $[a, b]$ such that $f (a)$ and $f (b)$ have
different signs.In each iteration, the function is evaluated in the middle point m between a and b, and sign of the function is different at a and m, and then the new interval $[a, b = m]$ is chosen for
the next iteration. Otherwise the interval $[a = m, b]$ is chosen for the next iteration.

### Newton's method

$$ y = f(x_0)+f'(x_0)(x_1-x_0)$$
$$ x_1 = x_0 - f(x_0)/f'(x_0)$$

```python
optimize.bisect(lambda x: np.exp(x) - 2, -2, 2)
```

```python
x_root_guess = 2
f = lambda x: np.exp(x)- 2
fprime = lambda x: np.exp(x)
optimize.newton(f, x_root_guess)
```

```python
optimize.newton(f, x_root_guess, fprime=fprime)
```

```python
optimize.brentq(lambda x: np.exp(x) - 2, -2, 2)
```

```python
optimize.brenth(lambda x: np.exp(x) - 2, -2, 2)
```

### Systems of Nonlinear Equations
The bisection method cannot be directly generalized to a multivariate equation system. On the other hand, Newton’s method can be used for multivariate problems.

$$ y-x^3-2x^2+1=0$$
$$y+x^2-1=0$$

```python
def f(x):
    return [x[1] - x[0]**3 - 2 * x[0]**2 + 1, 
            x[1] + x[0]**2 - 1]

optimize.fsolve(f, [1, 1])
```

```python
# Generate a Jacobian Matrix at first
x, y = sympy.symbols("x, y")
f_mat = sympy.Matrix([y - x**3 -2*x**2 + 1, 
                      y + x**2 - 1])
f_mat.jacobian(sympy.Matrix([x, y]))
```

```python
def f_jacobian(x):
    return [[-3*x[0]**2-4*x[0], 1], [2*x[0], 1]]
optimize.fsolve(f, [1, 1], fprime=f_jacobian)
```

The initial guess for the solution is important, and different initial guesses may result in different solutions.
