## Linear Equation Systems
$Ax =b$ where A is a m × n matrix, b is a m × 1 matrix (or m-vector), and
x is the unknown n × 1 solution matrix (or n-vector).

```python
from scipy import linalg as la
from scipy import optimize
import sympy
sympy.init_printing()
import numpy as np
import matplotlib.pyplot as plt
```

A system for which $n < m$ is said to be underdetermined (欠定), because it has fewer
equations than unknown and therefore cannot completely determine a unique solution.
If, on the other hand, $m > n$, then the equations are said to be overdetermined（超定）. This will
in general lead to conflicting constraints, resulting in that a solution does not exist.

$m = n$ is an important special case (unique solution). the
matrix $A$ must be nonsingular（非奇异）, in which case the inverse of $A$ exists, and the solution can
be written as $x = A^{−1}b$. If the matrix $A$ is singular, that is, the rank of the matrix （秩） is less than
$n$, $rank(A) < n$, or, equivalently, if its determinant（行列式） is zero, $detA = 0$, then the equation
$Ax = b$ can either have no solution or infinitely many solutions, depending on the right-hand
side vector $b$.

```python
A = sympy.Matrix([[2, 3], [5, 4]])
b = sympy.Matrix([4, 3])
A.rank()
```

```python
A.condition_number()
# 矩阵的条件数，衡量b的微小变化会导致x解的变化程度
# 条件数越接近1，方程组越良态，数值计算的精度
```

```python
sympy.N(_)
# 计算上一个式子的近似值
```

```python
A = np.array([[2, 3], [5, 4]])
b = np.array([4, 3])
np.linalg.matrix_rank(A)
```

```python
np.linalg.cond(A)
```

```python
# LUdecomposition
A = sympy.Matrix([[2, 3], [5, 4]])
b = sympy.Matrix([4, 3])
x = A.solve(b); x
```

```python
A = np.array([[2, 3], [5, 4]])
b = np.array([4, 3])
la.solve(A, b)
```
