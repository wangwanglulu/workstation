```python
import pandas as pd
db =pd.read_excel('db_top.xlsx')
```

```python
db
```

```python
genre = db[['中文名', '类型']].copy()
```

```python
genre
```

```python
all_g_list = []
for x in genre['类型']:
    all_g_list.append(x.split(','))
genre["类型"]=all_g_list
```

```python
genre
```

```python
x = pd.DataFrame()
x['类型'] = genre.iloc[0]['类型']
x["中文名"] = genre.iloc[0]['中文名']
x['d']=1
x= pd.pivot_table(x,index='中文名', columns='类型',values='d')
```

```python
x
```

```python
genre_list = pd.DataFrame()
for i in range(len(genre)):
    x = pd.DataFrame()
    x['类型'] = genre.iloc[i]['类型']
    x["中文名"] = genre.iloc[i]['中文名']
    x['d']=1
    x = pd.pivot_table(x,index='中文名', columns='类型',values='d')
    genre_list=pd.concat([genre_list,x])
g_l=genre_list.fillna(0)
```

```python
g_l.to_excel(excel_writer='g_l.xlsx', sheet_name="rank",\
                    encoding='utf-8')
```
